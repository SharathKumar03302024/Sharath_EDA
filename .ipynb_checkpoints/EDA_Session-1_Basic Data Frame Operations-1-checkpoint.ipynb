**Step-7**

$drop$ $the$ $Column$


- In order to drop the column 

- we need to use drop method

- It takes 3 parameters
        
    - drop column or row
        
    - mention the column name
        
    - axis
        
        - asix=1 refernce as column
           
        - axis=0 reference as row
            
     - you want to create a new data frame or 
         
     - you want to overwrite the existing dataframe
         
         - inplace=True 

```python
# provind the columns
data=zip(name,age)
cols=['Name','Age']
ind=['A','B','C']
city_names=['Hyd', 'blr','Chennai']
df=pd.DataFrame(data,columns=cols,index=ind)
df['city']=city_names
df
```

```python
df.drop('city',axis=1,inplace=True)
df
```

```python
df.drop('C',axis=0,inplace=True)
df
```

## Create Dataframes using dictionary##

```python
d1={"Name":['Ramesh','Suresh','Satish'],
   'AGE':[30,35,40]}

pd.DataFrame(d1)

# No need to zip
# No need of column names
```

```python
d1={"Name":['Ramesh','Suresh','Satish'],
   'AGE':[30,35,40]}
ind=['A','B','C']
pd.DataFrame(d1,index=ind)
```

```python
d1={"Name":['Ramesh','Suresh','Satish'],
   'AGE':[30,35,40]}
ind=['A','B','C']
city_names={"city":['Hyd1', 'blr','Chennai']}
d1.update(city_names) 
df=pd.DataFrame(d1,index=ind)
df
```

```python
city_names=['Sec', 'Santoshnagar','Lingampally']
df["city"]=city_names
df
```

```python
df.drop('city',axis=1,inplace=True)
```
