**One hot encoder**

- one hot encoder name says at a time one will On and other will Off

- For example case status has two lables 

    - Certified
    
    - Denied
    
- when you apply one hot encoding on case status , it creates two more extra columns 

    - case_status_certified
    
    - case_status_Denied
    
|case_status|case_status_Certified|case_status_denied
|-|-|-|
|Certified|
|Denied|

|case_status|case_status_Certified|case_status_denied
|-|-|-|
|Certified|1|0|
|Denied|0|1|
|Denied|0|1|
|Denied|0|1|
|Certified|1|0|

**Advantages**

- when you develop ML model it is very important the columns should be independent to each other

- so here case status creating two extra columns

- which are indepentdent each other, which means the row values at a time only one column has 1

- Columns are indepenent each other

- which means 90 degrees phase shift

- which means perpendicular each other

- which means orthoganal each other

**Disadvantage**

- The disadvantage is if a column has 100 unique lables , 100 new columns will be created 

- the data will become sparse, which means huge

- Columns are more means Dimentions are more

- the processing time is more 

- the memory consumption is more 

- **Curse of Dimensionality**
