## Create Dataframe using list ##

```python
UserName=['User1','User2','User3']
Salary=['1000','2000','4000']
UserName,Salary
```

```python
import pandas as pd
df=pd.DataFrame(zip(UserName,Salary))
df
```

```python
# provide column to DataFrame 
Data=zip(UserName,Salary)
pd.DataFrame(Data,columns=["Name",'Age'])
```

```python
# provide column to DataFrame  another example
Data=zip(UserName,Salary)
colm= ["Name",'Age']
pd.DataFrame(Data,columns=colm)
```

```python
# provide Index values
Data=zip(UserName,Salary)
colm= ["Name",'Age']
ind=['A','B','C']
pd.DataFrame(Data,columns=colm,index=ind)
```

```python
# add a new column to existing records
Data=zip(UserName,Salary)
colm= ["Name",'Age']
ind=['A','B','C']
df=pd.DataFrame(Data,columns=colm,index=ind)
df['Location']=['Loc1','Loc2','Loc3']
df
```

```python
# add a new column to existing records - Another example
Data=zip(UserName,Salary)
colm= ["Name",'Age']
ind=['A','B','C']
Loc=['Loc1','Loc2','Loc3']
df=pd.DataFrame(Data,columns=colm,index=ind)
df['Location']=Loc
df
```

```python
# update exsiting column 
df['Name']=['Ravi','Kishore','Ragav']
df
```

```python
df.drop("Age",axis=1,inplace=True)
```

```python
df
```

```python
df.drop('C',axis=0,inplace=True)
```

## Create Dataframes using dictionary##

```python
d1={"Name":['Ramesh','Suresh','Satish'],
   'AGE':[30,35,40]}

pd.DataFrame(d1)

# No need to zip
# No need of column names
```

```python
d1={"Name":['Ramesh','Suresh','Satish'],
   'AGE':[30,35,40]}
ind=['A','B','C']
pd.DataFrame(d1,index=ind)
```

```python
d1={"Name":['Ramesh','Suresh','Satish'],
   'AGE':[30,35,40]}
ind=['A','B','C']
city_names={"city":['Hyd1', 'blr','Chennai']}
d1.update(city_names) 
df=pd.DataFrame(d1,index=ind)
df
```

```python
city_names=['Sec', 'Santoshnagar','Lingampally']
df["city"]=city_names
df
```

```python
df.drop('city',axis=1,inplace=True)
```

```python
# drop row
df.drop('C',axis=0,inplace=True)
df
```
